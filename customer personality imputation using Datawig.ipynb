## Datawig for imputation
We will use datawig library which is a framework used impute missing values in tables.<br>
Datawig is a Deep Learning library developed by AWS Labs and is primarily used for missing values imputation.<br>
DataWig currently works with only python version 3.7 and below.

```python
from datawig.utils import random_split
from datawig import SimpleImputer
from sklearn.metrics import mean_absolute_error, mean_squared_error

```

```python
# Split the rows containing no null values into training and test data.
df_train, df_test = random_split(df.dropna(axis=0), split_ratios=[0.8, 0.2])
input_cols = ['Education','Marital_Status','MntWines','MntFruits', 'MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds','Total_Spent','children_home','NumDealsPurchases','Total_Num_of_Purchases','NumWebVisitsMonth','NumStorePurchases','NumCatalogPurchases','Income_categories']
output_Income = 'Income' # Numerical output
```

```python
# Initialize imputer for numerical imputation
imputer_Income = SimpleImputer(
                 input_columns=input_cols,
                 output_column=output_Income,
                 output_path='imputer_Income'  # Column to be imputed
                 )

```

```python
# Imputation model fit
imputer_Income.fit(train_df=df_train) 
```

```python
# Predict the income for the test data.
predictions_Income = imputer_Income.predict(df_test)

```

```python
predictions_Income.head()
```

```python
mae_income = mean_absolute_error(df_test[output_Income],
                                 predictions_Income[f'{output_Income}_imputed'])


print('The mean absolute error for the Income predicted: {}'.format(mae_income))
```

```python
missing_income = df[df['Income'].isna()]
```

```python
predicted_income = imputer_Income.predict(missing_income)
df=df.merge(predicted_income, how='left')

```

```python
df['Income'].fillna(df['Income_imputed'] ,inplace=True)
```

```python
df.drop(columns='Income_imputed', inplace=True)
```

```python
df.to_csv('marketing_campaign_imputed.csv', index=False)
```
